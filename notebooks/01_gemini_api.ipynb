# Using Gemini API

First, you need to go to [Google AI Studio](https://aistudio.google.com/app/api-keys) and create your API key.

Now, we will learn some principles of connecting to the Gemini API.

## Libs

```python
import os

from google import genai
from google.genai import types
from dotenv import load_dotenv
from argparse import ArgumentParser, Namespace
```

## Creating a client

Create a `.env` file and add name your api key as `GEMINI_API_KEY`. After that, we will create a client object with `genai.Client()`.

```python
# Loads the env variables
load_dotenv()

# Gets the variable GEMINI_API_KEY
api_key = os.getenv("GEMINI_API_KEY")

# Creates a client object
client = genai.Client(api_key=api_key)

# Creating user prompt
user_prompt = "Hello there, what is a good first move in chess while playing whites?"

# Calling Gemini for answering a prompt
response = client.models.generate_content(model="gemini-2.5-flash", contents=user_prompt)
```

The `generate_content()` method returns a [`GenerateContentResponse`](https://docs.cloud.google.com/vertex-ai/generative-ai/docs/reference/rest/v1/GenerateContentResponse) object. This object has some fields that we will use:

- `text`: The text response from the model.
- `usage_metadata`: Has `prompt_token_count` and `candidates_token_count` properties.

```python
print(f"Prompt tokens: {response.usage_metadata.prompt_token_count}")
print(f"Response tokens: {response.usage_metadata.candidates_token_count}")
print(f"Model response: \n{response.text}")
```

## Multiple messages

We can also pass multiple messages to the `contents` parameter. For that purpose, we can use the `types` module.

```python
messages = [
    types.Content(role="user", parts=[types.Part(text=user_prompt)])
]

response = client.models.generate_content(model="gemini-2.5-flash", contents=messages)

print(f"Prompt tokens: {response.usage_metadata.prompt_token_count}")
print(f"Response tokens: {response.usage_metadata.candidates_token_count}")
print(f"Model response: \n{response.text}")
```

## Command line arguments

For scripts, we will use command line arguments for the user prompt and the verbosity of the response. So, let's create a function that parses arguments from the command line.

```python
def parse_args() -> Namespace:
    """
    Parses all arguments passed in the command line.

    Returns:
        Namespace: Object containing all arguments from the command line.
    """
    # Creating object
    parser = ArgumentParser(description="Chatbot")
    
    # Adding argument for the user prompt
    parser.add_argument("user_prompt", type=str, help="User prompt")
    
    # Adding argument for the verbosity
    parser.add_argument("--verbose", action="store_true", help="Enable verbose output")
    
    # Places extracted data in Namespace object
    args = parser.parse_args()
    
    return args
```

With this function we can now access the arguments in the object's fields:

- `user_prompt`: Stores user prompt.
- `verbose`: Stores values `True` of `False`.
