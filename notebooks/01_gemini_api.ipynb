# Using Gemini API

First, you need to go to [Google AI Studio](https://aistudio.google.com/app/api-keys) and create your API key.

Now, we will learn some principles of connecting to the Gemini API.

## Libs

```python
import os

from google import genai
from google.genai import types
from dotenv import load_dotenv
```

## Creating a client

Create a `.env` file and add name your api key as `GEMINI_API_KEY`. After that, we will create a client object with `genai.Client()`.

```python
# Loads the env variables
load_dotenv()

# Gets the variable GEMINI_API_KEY
api_key = os.getenv("GEMINI_API_KEY")

# Creates a client object
client = genai.Client(api_key=api_key)

# Calling Gemini for answering a prompt
response = client.models.generate_content(model="gemini-2.5-flash", contents="Hello there, what is a good first move in chess while playing whites?")
```

The `generate_content()` method returns a `GenerateContentResponse` object. This object has some fields that we will use:

- `text`: The text response from the model.
- `usage_metadata`: Has `prompt_token_count` and `candidates_token_count` properties.

```python
print(f"Prompt tokens: {response.usage_metadata.prompt_token_count}")
print(f"Response tokens: {response.usage_metadata.candidates_token_count}")
print(f"Model response: \n{response.text}")
```
